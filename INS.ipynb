# Feature Extraction

```python
from scipy.stats import pearsonr

#Person Correlation Calculation

#List of features to check against target

selected_features= [
    'age', 'bmi', 'children', 'is_female', 'is_smoker',
    'region_northwest', 'region_southeast', 'region_southwest',
    'bmi_category_Normal', 'bmi_category_Overweight', 'bmi_category_Obese',
]
correlations = {
    feature: pearsonr(df_clean[feature], df_clean['charges'])[0]
    for feature in selected_features
}
correlation_df = pd.DataFrame(list(correlations.items()), columns=['feature', 'Pearsonr Correlation'])
correlation_df.sort_values(by='Pearsonr Correlation', ascending=False)
```

```python
print(df_clean.columns.tolist())
```

```python
cat_features = [
    'is_female','is_smoker'
    ,'region_northwest','region_southeast'
    ,'region_southwest','bmi_category_Normal'
    ,'bmi_category_Overweight','bmi_category_Obese'
]
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

alpha = 0.05

df_clean['charges_bin']=pd.qcut(df_clean['charges'],q=4,labels=False)
chi2_results={}
for col in cat_features:
    contingency=pd.crosstab(df_clean[col], df_clean['charges_bin'])
    chi2_stat, p_val, _, _ =chi2_contingency(contingency)
    decision = 'Reject Null(Keep Feature)' if p_val < alpha else 'Accept Null(Keep Feature)'
    chi2_results[col]= {
        'chi2_statistic': chi2_stat,
        'p_value': p_val,
        'decision': decision
    }
chi2_df = pd.DataFrame(chi2_results).T
chi2_df = chi2_df.sort_values(by='p_value')
chi2_df
```

```python
final_df = df_clean[['age', 'is_female', 'bmi', 'children', 'is_smoker', 'charges', 'region_southeast', 'bmi_category_Obese']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_df.drop('charges', axis=1)
y = final_df['charges']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
from sklearn.linear_model import  LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
r2

n = X_test.shape[0]
p = X_test.shape[1]

adjusted_r2 = 1-((1-r2)*(n-1)/(n-p-1))
adjusted_r2
```
